```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import norm
```

### Plotting a function

```python
x = np.linspace(-4, 4, 50)
y = x**2
plt.plot(x, y)
```

### Plotting the normal distribution

```python
x = np.linspace(-3, 3, 50)
y = norm.pdf(x, loc=0, scale=1)
plt.plot(x,y)
```

### Writing to a text file

```python
f = open("test.txt", "w")
f.write("This is just a test.")
f.close()
```

```python
f = open("test.txt", "a")
f.write("\nThis is a second test.")
f.close()
```

```python
with open("test.txt", "a") as f:
    f.write("\nThis is a third test.")
```

```python
dct = dict(a=[1, 2, 3], b=[4, 5, 6])
df = pd.DataFrame(dct)
with open("test.txt", "a") as f:
    f.write("\n\n" + df.to_latex(index=False))
```

### Reading a text file

```python
with open("test.txt", "r") as f:
    lines = f.readlines()

for line in lines:
    print(line)
```
